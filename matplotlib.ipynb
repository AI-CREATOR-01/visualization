```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# yah function 1 se 20 ke bich me 10 data generate krdega jo ki ek 1D array hoga
x = np.linspace(1,20,10)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('this is my data')
plt.xlabel('this is my x data')
plt.ylabel('this is my y data')
plt.plot(x,y,)
plt.show()
```

```python
a = np.random.rand(50)
b = np.random.rand(50)
```

```python
# a,b represent the x and y axis data,
# c ---> represent the color of dots we want
#  s ---> represent the size of dots that we want
# alpha ---->> represent the intensity of color
# marker ---->> represent the shape of dot
plt.scatter(a,b, c = 'b', s= 100, alpha = 0.1, marker = 'v')
```

```python
data1= [1,2,3,4,1,2,3,4,6,7,8,8,7,6]
```

```python
data2 = [1,2,3,4,5]
data3 = [6,7,8,9,10]
```

```python
plt.scatter(data2,data3)


# if we want that each dot should be different in color the we can do this also see
# we have total 5 data on x axis and y axis so define five different color as list see
data4 = ['b','r','w','g','y']
plt.scatter(data2,data3, c= data4)
```

```python
x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)
x,y,z
```
