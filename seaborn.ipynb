```python
import seaborn as sb
```

```python
# yah data phle se hi is library6 me hota hai iris naam kaa
# yah ek flower hota hai 
iris = sb.load_dataset('iris')
```

```python
iris
```

```python
# let see the relation bw sepal length and sepal width see

sb.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
# displot equivalent to hist() which represent the data frequency on y axis and 
# range od data on x axis
sb.displot(iris['sepal_length'])
```

```python
tips = sb.load_dataset('tips')
```

```python
tips
```

```python
sb.scatterplot(x = tips.total_bill, y = tips.tip)
# so by graph  we can say that jisne sbse jyada bill pay kiya hai usne,
# sbse jyada tip bhi di hai mtlb overall hmne tips vs bills ka relation nikala hai 
```

```python
# jo smoker column hai usme do hi types ka data tha ek no dusra yes mtlb yah function,
# kisi bhi column ke total values ko count krta hai
tips['smoker'].value_counts()
```

```python
#  mtlb NO category waale total 151 hai aur YES category waale total 93 hai
```

```python
# but hm chahte hai ki jo relation bill vs tip ka nikala hai usi same graph se hi,
# smoker waale column/data ko bhi analysis kr skte hai ki smoker me kaunsi category jyada matra me hai
# with the help of HUE/STYLE
# HUE------->> yah dots ko diifferent color me bant deta hai
# STYLE------->>>> yah dots ko different shape me baant deta hai to represent different category/value ,
# of a column see below 
# data = tips yah likhkar hmne complete tips dataset ko mentioned kiya hai tabhi hue ya style attribute
# work krenge
sb.scatterplot(x = tips.total_bill, y = tips.tip, data =tips, hue = 'smoker', style ='sex' )
```

```python
# NOTE---------->>>>>>>>>>

# UPPER HMNE EK GRAPH PLOT KIYA TOTAL_BILL VS TIP TAAKI HM EN DONO KE BICH KA RELATION KO JAAN SKE,
# AUR ANALYSIS KR SKE JAISE KI BIILS BAD RHA HAI TO KYA TIPS BHI INCREASE HO RHA HAI YA NHI SO HO RHA HAI

# BUT LGE HAATH HMNE SEX AND SMOKER WAALE COLUMN KE DATA KO BHI ANALYSIS KRLIYA JAISE YELLOW WAALE DOTS,
# JYADA NAZAR AA RHE HAI AUR YELLOW NO CATEGORY KO REPRESENT KR RHA HAI SO JO COSTUMER SMOKE NHI KRTE HAI ,
# WO JYADA MATRA ME HAI 

# CIRCULAR DOTS SEX KE MALE CATEGORY KO REPRESENT KR RHA HAI AUR CIRCLE DOTS JYADA HAI MTLB MALE CUSTOMER ,
# JYADA THE 

# JO SMOKER THA USNE JYADA TIPS AUR BILLS PAY KIYA THA HM YE BHI BTA POAA RHE N HAI NICE ANALYSIS 
```

```python
sb.scatterplot(x=tips.total_bill,y=tips.tip, data = tips, style = 'size')
```

```python
# so by obeserving the graph we can say that jo size column hai usme total catregory 6 hai mtlb is pure,
# column me total 6 type ke data/value nazar aayenge
# now square shape 3 ko represent kr rha hai mtlb jo 3 ke group me aaye the unhone sbse jyada bill and tip,
# pay kiye hai
# cross represent 2 mtlb wah customere jo 2 ke saathb the (including itself) unki sankhya jyada thi so,
# hm kuchh aise analysis kr skte hai of any column in same graph done.
```

```python
sb.catplot(x='smoker', y = 'tip', data = tips)
```

```python
# data = tips likhar hm ye bta rhe hai x and y axis pr jo column/data lena hai wah yha se uthana hai

# yah function histograph ko bhi add krdeta hai taaki lge hath range bhi pta kr paaye ki kis range kya
# frequency hai
sb.jointplot(x='total_bill',y = 'tip',data = tips)
```

```python
# yah function tips naam ke data me se saare numerical column ko represent krdega taaki kisi bhi do data
# ka relation pta kr paaye
sb.pairplot(tips)
```
